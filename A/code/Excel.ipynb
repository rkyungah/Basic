### 여러 개의 엑셀 파일에서 데이터 수정

```python
file_name = './data/담당자별_판매실적_윤종필사원.xlsx'

new_file_name = re.sub(".xlsx","2.xlsx", file_name)
new_file_name
```

```python
# 원하는 문자열이 포함된 파일을 검색해 리스트 할당

excel_data_files = glob.glob('data/담당자별_판매실적_*.xlsx')
excel_data_files
```

```python
#리스트에 있는 엑셀 파일만큼 반ㄴ복

for i in excel_data_files :
    # 엑셀 파일에서 데이터프레임 형식으로 데이터 가져오기
    df = pd.read_excel(f)
    
    #특정 열의 값을 변경
    if (df.loc[1,'담당자'] =='윤종필') :
        df['담당자'] = '윤종필사원'
    elif (df.loc[1,'담당자']) == '홍성민' :
        df['담당자'] = '홍성민사원'
    elif (df.loc[1,'담당자']) == '신은영' :
        df['담당자'] = '신은영사원'
    elif (df.loc[1,'담당자']) == '김재경' :
        df['담당자'] = '김재경사원'
        
    # 엑셀 파일 이름에서 지정된 문자열 패턴을 찾아서 파일명을 변경
    
    f_new = re.sub('.xlsx','_v2.xlsx', f)
    print(f_new)
    
    #수정된 데이터를 새로운 이름의 엑셀 파일로 저장
    new_excel_file = pd.ExcelWriter(f_new, engine = 'xlsxwriter')
    df.to_excel(new_excel_file, index=False)
    new_excel_file.save()
```

```python
df = pd.read_excel(excel_file_name)

# 제품명 열에서 아이폰 XR이 있는 행만 선택
iPhone=df[(df['제품명'] == '아이폰XR')]

# 행 별로 합계를 구하고 마지막 열 다음에 추가
iPhone_sum = pd.DataFrame(iPhone.sum(axis=1), columns = ['연간판매량'])
iPhone_total = iPhone.join(iPhone_sum)

# 열별로 합해 분기별 합계와 연간판매량 합계를 구하고 마지막 행 다음에 추가한다.
iPhone_sum2 = pd.DataFrame(iPhone_total.sum(axis=0), columns = ['합계'])
iPhone_total2 = iPhone_total.append(iPhone_sum2.T)

# 지정된 항목의 문자열을 변경한다.
iPhone_total2.loc['합계','제품명'] = '아이폰XR'
iPhone_total2.loc['합계','담당자'] = '전체'
iPhone_total2.loc['합계','지역'] = '전체'

#결과 확인
iPhone_total2
```

```python
labels = 'Frogs', 'Hogs','Dogs', 'Cats'
sizes=[15,30,45,10]
explode = (0,0.3,0,0)    #몇 번째를 얼마나 띄워 놓을지.

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
       shadow=True,startangle=90)
ax1.axis('equal')       #원이 타원이 되지 않도록 가로세로 비율을 똑같이

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,12), subplot_kw=dict(aspect="equal"))

recipe = ["375 g 밀가루",
         "75 g 설탕",
         "250 g 버터",
         "300 g 딸기"]

data = [float(x.split()[0]) for x in recipe]   # recipe의 맨 앞글자
ingredients = [x.split()[-1] for x in recipe]  # recipe의 뒷글자

def func(pct, allvals) :
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}%\n({:d}g)".format(pct, absolute)

# {:.1f} = 소숫점 한자리수까지 -> {:.2f} = 소숫점 두자리수까지
#"{:.1f}%   = 소수점 한자리수까지%
#\n 한 칸 띄우기
# ({:d}g)   = 수치 g

wedges, texts, autotexts = ax.pie(data, autopct=lambda pct : func(pct,data),
                                 textprops = dict(color='w'))

ax.legend(wedges, ingredients,
         title="주요성분",
         loc="center left",
         bbox_to_anchor=(1,0,0.5,1))

plt.setp(autotexts, size=14, weight="bold")

ax.set_title("빵의 주요 성분")

plt.show()

```

## box 그래프 작성

```python
# scale : 분산
# loc : 평균값

s1 = np.random.normal(loc=0,  scale=1.0, size=500)    # loc : Mean ("centre") of the distribution.
s2 = np.random.normal(loc=5,  scale=0.5, size=500)    # scale : Standard deviation (spread or "width") of the distribution.
s3 = np.random.normal(loc=10, scale=2.0, size=500)    # size : Output shape
```

```python
plt.figure(figsize=(10,4))
plt.plot(s1, label='s1')
plt.plot(s2, label='s2')
plt.plot(s3, label='s3')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.boxplot((s1, s2, s3))
plt.grid()
plt.show()
```

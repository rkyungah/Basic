## lmplot  (대문자 i아님)
> 변수 간 상관관계를 표시할 때 자주 사용

```python
sns.set_style('whitegrid')
tips = sns.load_dataset('tips')
tips.head(5)
```

```python
sns.set(color_codes=True)
sns.lmplot(x='total_bill',y='tip',data=tips)
plt.show()
```

```python
sns.lmplot(x="total_bill", y="tip", hue='smoker',data=tips)  
plt.show()
```

```python
sns.lmplot(x="total_bill", y="tip", hue='smoker',data=tips,
          markers=["o","x"])
plt.show()
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips, palette="Set1",
          markers=[">","<"])
plt.show()
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips,
           palette=dict(Yes="r", No="m"))
plt.show()
```

```python
sns.lmplot(x="total_bill", y="tip", col="smoker", data=tips)
plt.show()
```

```python
sns.lmplot(x='size',y='total_bill',hue='day',col='day',data=tips,aspect=0.4,
           x_jitter=.1,palette="Set1")
plt.show()
```

```python
sns.lmplot(x="total_bill", y="tip", col="day", hue="day", data=tips,
           col_wrap=4, size=6,aspect=0.4)
plt.show()
```

```python
sns.lmplot(x="total_bill", y="tip", row="sex", col="time", data=tips, size=3)
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',row='sex',col='time',data=tips,
          size=3)
g = (g.set_axis_labels('total_bill(US Dollars)','Tip')).set(xlim=(0,60),
                                                           ylim=(0,12),
                                                            xticks=[10,30,50],
                                                            yticks=[2,6,10]).fig.subplots_adjust(wspace=.02)
plt.show()
```

## kdeplot

```python
np.random.seed(10)
```

```python
mean, cov = [0,2],[(1, .5),(.5, 1)]
```

```python
x,y = np.random.multivariate_normal(mean,cov,size=50).T
x
```

```python
y
```

```python
plt.plot(x)
plt.show()
```

```python
plt.plot(y)
plt.show()
```

```python
sns.kdeplot(x)
plt.show()
```

```python
sns.kdeplot(x, shade=True, color='r')
plt.show()
```

```python
sns.kdeplot(x,y)
plt.show()
```

```python
sns.kdeplot(x,y,shade=True)
plt.show()
```

```python
sns.kdeplot(x, y, n_levels=50, cmap='Purples_d')
plt.show()
```

```python
sns.kdeplot(x,bw=.15,)
plt.show()
```

```python
sns.kdeplot(y, vertical=True)   # vertical : 뒤집기
plt.show()
```

```python
sns.kdeplot(x, bw=.15, cut=2)
plt.show()
```

```python
iris = sns.load_dataset("iris")
setosa = iris.loc[iris.species == "setosa"]
virginica = iris.loc[iris.species == "virginica"]
sns.kdeplot(setosa.sepal_width, setosa.sepal_length, cmap="Reds",
            shade=True, shade_lowest=False)
sns.kdeplot(virginica.sepal_width, virginica.sepal_length,
            cmap="Blues", shade=True, shade_lowest=False)
plt.show()
```

## distplot
> http://seaborn.pydata.org/generated/seaborn.distplot.html?highlight=distplot#seaborn.distplot

```python
sns.set(rc={"figure.figsize": (8, 4)}); np.random.seed(0)
x = np.random.randn(100)
sns.distplot(x)
plt.show()
```

```python
x = pd.Series(x, name="x variable")
sns.distplot(x)
plt.show()
```

```python
sns.distplot(x, rug=True, hist=False)   # rug : 바코드 모양
plt.show()
```

```python
sns.distplot(x, rug=True, hist=True)
plt.show()
```

```python
from scipy.stats import norm
```

```python
sns.distplot(x, rug = True, fit = norm, kde=False)
plt.show()
```

```python
sns.distplot(x, rug = False, fit = norm, kde=True, vertical=True)
plt.show()
```

```python
sns.set_color_codes()
sns.distplot(x, color="y")
plt.show()
```

```python
sns.distplot(x, rug=True, rug_kws={"color": "g"},
             kde_kws={"color": "k", "lw": 3, "label": "KDE"},
             hist_kws={"histtype": "step", "linewidth": 3, 
                       "alpha": .5, "color": "g"})
plt.show()
```

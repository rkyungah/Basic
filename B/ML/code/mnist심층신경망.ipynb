## **01 신경망 모델 구성하기**


*   MNIST는 28 * 28로 구성되어 있다.
*   784개의 특징(픽셀)로 구성되어 있다.
*   레이블은 0~9까지 10개의 분류


```
# 우리의 신경망
## 입력층 - 784 노드(unit)
## 은닉층 (3개층) - 64노드(unit)
## 출력층 (범주 10개) - 10노드(unit)
```

```
X = tf.placeholder(tf.float32, [None,784])
Y = tf.placeholder(tf.float32, [None,10])
```

```
W1 = tf.Variable(tf.random_normal([784, 64], stddev = 0.01)) # stddev = 표준편차
L1 = tf.nn.relu(tf.matmul(X,W1))

W2 = tf.Variable(tf.random_normal([64, 64], stddev = 0.01))
L2 = tf.nn.relu(tf.matmul(L1, W2))

W3 = tf.Variable(tf.random_normal([64, 64], stddev = 0.01))
L3 = tf.nn.relu(tf.matmul(L2,W3))

W4 = tf.Variable(tf.random_normal([64, 10], stddev= 0.01))
model = tf.matmul(L3,W4)

print(model)
```

### **비용함수, 최적화 함수 지정**

*  AdamOptimizer은 RMSProp와 Momentum방식의 합

```
cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits=model,
                                                                labels=Y))
## AdamOptimizer(learning_rate)
optimizer = tf.train.AdamOptimizer(0.001).minimize(cost)
```

```
init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)
```

```
total_cost = 10
for i in range(55) :
  batch_xs, batch_ys = mnist.train.next_batch(1000) # x, y label을 1000개씩 넘겨준다
  _, cost_val = sess.run([optimizer, cost], feed_dict={X : batch_xs, Y : batch_ys})
  #학습
  
  # 총 loss의 계산
  # loss의 합 : cost
  # 55번 돌아가면서 발생된 cost
  total_cost = total_cost + cost_val
  print("한 번 학습 시의 cost", cost_val)
  
print("1 Epoch 평균 cost : {:.3f}".format(total_cost/60))
```

```
batch_size = 100

for j in range(1, 31) :
  total_cost = 0
  for i in range(batch_size) :
    batch_xs, batch_ys = mnist.train.next_batch(batch_size) # x, y label을 1000개씩 넘겨준다
    _, cost_val = sess.run([optimizer, cost], feed_dict={X : batch_xs, Y : batch_ys})
    #학습

    # 총 loss의 계산
    # loss의 합 : cost
    # 55번 돌아가면서 발생된 cost
    total_cost = total_cost + cost_val
  print("Epoch {} \t 평균 cost : {}".format(j,total_cost/batch_size))
```
